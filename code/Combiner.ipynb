## Combine CSVs

Some of the CSVs have columns that are not required for the final choropleth visualization. As we are going through them, we should drop and rename columns to make it easier for the final choropleth to work! We will be merging our dataframes based on year, month and violation precinct.

### Initialize Master (Output) Dataframe

```python
# The car_types_max_percentage_names has all the required rows, which we can use to initialize our master dataframe
df = pd.read_csv("../data/final_data_to_join/car_types_max_percentage_names.csv", index_col = 0)

# Let's name some of the columns so that they are more recognizable even after merging all dataframes
df = df.rename(columns={"Max Percentage Column": "maximum_ticketed_car_type", "Max Percentage Value": "percentage_maximum_ticketed_car_type"})
df = df.round({'percentage_maximum_ticketed_car_type': 0})
df = df.astype({'percentage_maximum_ticketed_car_type': 'int'})

# Initialize master df
master_df = df
```

```python
display(master_df)
```

### Combine with color_data.csv

```python
# Read
df = pd.read_csv("../data/final_data_to_join/color_data.csv")

# We only need vehicle color and percent_maxcolor from this df
df = df.drop(columns = ['total_maxcolor', 'total_allcolors'])

# We can also round percent_maxcolor to increase ease of understanding for users
df = df.round({'percent_maxcolor': 0})
df = df.astype({'percent_maxcolor': 'int'})

# Let's rename into more recognizable form
df = df.rename(columns={"Vehicle Color": "maximum_ticketed_vehicle_color", "percent_maxcolor": "percentage_maximum_ticketed_vehicle_color"})
```

```python
# Let's merge it with master_df
master_df = pd.merge(master_df, df, how = 'outer', on = ['Year', 'Month', 'Violation Precinct'])
```

```python
# Let's merge it with master_df
master_df = pd.merge(master_df, df, how = 'outer', on = ['Year', 'Month', 'Violation Precinct'])
```

### Combine with plate_type_data.csv

```python
# Read
df = pd.read_csv("../data/final_data_to_join/plate_type_data.csv")

# Make new columns for saying which is higher, passenger or commercial and if higher, what is percentage
df["maximum_ticketed_plate_type"] = np.where(df["percent_passenger_vehicles"] > df["percent_commercial_vehicles"], "passenger", "commercial")
df["percent_maximum_ticketed_plate_type"] = np.where(df["maximum_ticketed_plate_type"] == "passenger", df["percent_passenger_vehicles"], df["percent_commercial_vehicles"])
df = df.round({'percent_maximum_ticketed_plate_type': 0})
df = df.astype({'percent_maximum_ticketed_plate_type': 'int'})

# Drop extra columns
df = df.drop(columns = ['total_passenger_vehicles', 'total_commercial_vehicles', 'total_other_vehicles', 'total_vehicles', 'percent_passenger_vehicles', 'percent_commercial_vehicles', 'percent_other_vehicles'])
```

```python
# Let's merge it with master_df
master_df = pd.merge(master_df, df, how = 'outer', on = ['Year', 'Month', 'Violation Precinct'])
```

### Combine with potential_quotas.csv

```python
# Read
df = pd.read_csv("../data/final_data_to_join/potential_quotas.csv")

# Just rename required here
df = df.rename(columns = {'Last 5 Days Meet Condition': 'quotas_condition_met'})
```

```python
# Let's merge it with master_df
master_df = pd.merge(master_df, df, how = 'outer', on = ['Year', 'Month', 'Violation Precinct'])
```

### Combine with time_of_day_summary.csv

```python
# Read
df = pd.read_csv("../data/final_data_to_join/time_of_day_summary.csv")

# Drop the extra column
# df = df.drop(columns = ['Unnamed: 4'])

# Rename
df = df.rename(columns = {'Precinct':'Violation Precinct','Time Most Tickets Are Issued':'max_tickets_issued_time_of_day'})
```

```python
# Let's merge it with master_df
master_df = pd.merge(master_df, df, how = 'outer', on = ['Year', 'Month', 'Violation Precinct'])
```

### Finally, combining with collisions.csv

```python
# Read
df = pd.read_csv("../data/final_data_to_join/collisions.csv", index_col = 0)

# Just rename cols here
df = df.rename(columns = {'Precinct': 'Violation Precinct', 'Collisions':'collisions'})
df.dtypes
```

So now we can change time column and output it to be the final data!
