Collisions data are collected from nyc.gov (https://www.nyc.gov/site/nypd/stats/traffic-data/traffic-data-archive.page)

The below code reads, cleans and combines the collisions data per precinct/month/year.

NOTE: \
As we couldn't create a sample for the collisions data, this notebook won't work.
This notebook essentially loads all the collisions data and groups it by year, month, precinct,
finally outputting collisions.csv. Data can be downloaded from the link above, if this file needs to be tested.

```python
import pandas as pd
import os
import zipfile

data_folder_path = '../../data/collisions_data'
zip_files = [file for file in os.listdir(data_folder_path) if file.lower().endswith('.zip')]

final_df = pd.DataFrame()

for zip_file in zip_files:
    try:
        extracted_folder_name = os.path.splitext(zip_file)[0]
        extracted_folder_path = os.path.join(data_folder_path, extracted_folder_name)

        zip_file_path = os.path.join(data_folder_path, zip_file)
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extracted_folder_path)

        excel_files = [file for file in os.listdir(extracted_folder_path) if file.lower().endswith(('.xlsx', '.xls')) and 'city' in file.lower()]
        if not excel_files:
            print(f"No matching Excel file found in {extracted_folder_name}")
            continue

        file_path = os.path.join(extracted_folder_path, excel_files[0])
        
        if extracted_folder_name[:4] == '2014':
            city_acc_df = pd.read_excel(file_path, header=4)
        else:
            city_acc_df = pd.read_excel(file_path, header=3)
        city_acc_df = city_acc_df.rename(columns={'GeoCode': 'Precinct', 'Number_of_Motor_Vehicle_Collisions': 'Collisions'})
        city_acc_df = city_acc_df[['Precinct', 'Collisions']]
        city_acc_df['Precinct'] = pd.to_numeric(city_acc_df['Precinct'], errors='coerce', downcast='integer')
        city_acc_df['Collisions'] = pd.to_numeric(city_acc_df['Collisions'], errors='coerce', downcast='integer')
        city_acc_df = city_acc_df.dropna(subset=['Precinct', 'Collisions'])
        city_acc_df['Year'] = int(extracted_folder_name[:4])
        city_acc_df['Month'] = int(extracted_folder_name[5:7])
        city_acc_df = city_acc_df.astype({'Precinct': 'int', 'Collisions': 'int', 'Year': 'int', 'Month': 'int'})

        final_df = pd.concat([final_df, city_acc_df])
        print(f"Processed {zip_file}")
    except Exception as e:
        print(f"Error processing {zip_file}: {str(e)}")
        continue

final_df = final_df.reset_index(drop=True)
final_df



```

```python
final_df.to_csv('collisions.csv')
```
