Removing some more features to make the dataset smaller 
-> Making it easier to load and process

```python
import pandas as pd
```

```python
# spriha - time of day analysis

year_list = [i for i in range(2014, 2024)]

for year in year_list:

    print(f"Started {year}")
    
    # load data
    data = pd.read_csv(f"../../data/main_data/cleaned/Parking_Violations_Issued_{str(year)}.csv",
                       dtype="string")
    
    data2 = data[["Issue Date", "Violation Precinct", "Violation Time"]]

    # save data
    data2.to_csv(f"../../data/main_data/time_of_day/Parking_Violations_Issued_{str(year)}.csv", index=False)

    print(f"Finished {str(year)}")
```

```python
# divij - color and plate type analysis

year_list = [i for i in range(2014, 2024)]

for year in year_list:

    print(f"Started {year}")
    
    # load data
    data = pd.read_csv(f"../../data/main_data/cleaned/Parking_Violations_Issued_{str(year)}.csv",
                       dtype="string")
    
    data2 = data[["Issue Date", "Violation Precinct", "Vehicle Color", "Plate Type"]]

    # save data
    data2.to_csv(f"../../data/main_data/color_and_plate_type/Parking_Violations_Issued_{str(year)}.csv", index=False)

    print(f"Finished {str(year)}")
```
